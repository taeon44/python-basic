#### 001 선 그래프

- 다음의 그림과 같이 plot을 완성하세요.
- 사용된 데이터 범위는 flights 데이터프레임의 passengers 컬럼 시리즈 중 0~11의 인덱스 데이터를 사용했습니다.

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
plt.title('flights')
plt.plot(range(1, 13), flights.loc[0:11,'passengers'],
        c='k', ls=':', marker='o', mec='b', mfc='b')
plt.ylabel('sepal_length')
plt.show()
````

#### 002 다중 그래프

- 다음의 그림과 같이 plot을 완성하세요.
- 사용된 데이터 범위는 flights 데이터프레임의 passengers 컬럼의 시리즈 중 0 ~ 11의 인덱스 데이터와 12 ~ 23 인덱스 데이터를 사용했습니다.

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
plt.subplot(2,1,1)
plt.title('flights')
plt.plot(range(1,13), flights.loc[0:11,'passengers'],
        c='g', ls='-.', marker='s')
plt.ylabel('sepal_length')


plt.subplot(2,1,2)
plt.plot(range(1,13), flights.loc[12:23,'passengers'],
        c='k', ls='-.', marker='s')
plt.ylabel('sepal_length')

plt.show()
````

#### 005 파이차트

- titanic 데이터에서 남녀의 수를 구하세요.

#### 007 산점도

-------
<span style="font-weight:normal; color:orange;">Print</span>

#### 012 heatmap

- titanic 데이터셋을 이용해 상관관계를 분석 후, 다음의 그림과 같이 heatmap 을 완성하세요

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
df = titanic.corr()
sns.heatmap(data=df, annot=True, cmap='crest')
plt.title("heatmap")
plt.show()
````

#### 013 violinplot
